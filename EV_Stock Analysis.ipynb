Daily percentage 


```python
tesla['Returns'] = (tesla['Close']/tesla['Close'].shift(1)) - 1
```

```python
nissan['Returns'] = (nissan['Close']/nissan['Close'].shift(1)) - 1
volkswagen['Returns'] = (volkswagen['Close']/volkswagen['Close'].shift(1)) - 1
```

```python
plt.plot(tesla['Returns'])
plt.plot(nissan['Returns'])
plt.plot(volkswagen['Returns'])
```

```python
tesla['Returns'].hist(bins=50)
```

```python
tesla['Returns'].hist(bins=100,label='Tesla',alpha=0.6)
nissan['Returns'].hist(bins=100,label='Nissan',alpha=0.6)
volkswagen['Returns'].hist(bins=100,label='volkswagen',alpha=0.6)

plt.legend()
plt.savefig('invest.png',bbox_inches='tight')
plt.show()
```

```python
tesla['Returns'].plot(kind='kde',label='Tesla')
nissan['Returns'].plot(kind='kde',label='Nissan')
volkswagen['Returns'].plot(kind='kde',label='Volkswagen')

plt.legend()
plt.savefig('invest1.png',bbox_inches='tight')
plt.show()
```
