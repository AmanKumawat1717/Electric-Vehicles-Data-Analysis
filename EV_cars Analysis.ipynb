```python

def color_generate():
    s=list("1234567890abcdef")
    val="".join(np.random.choice(s,6))
    return "#" + val
```

```python
plt.rcParams['figure.figsize']=11,5
plt.rcParams['figure.dpi']=120
#plt.style.use('classic')
fig=plt.figure()
ax=plt.axes()
x=data['Brand'].value_counts().index
y=data['Brand'].value_counts().values
ax.bar(x, y, color=[color_generate() for i in range(len(x))])
plt.xticks(rotation=90)
plt.xlabel("Brand")
plt.ylabel("Model")
plt.title("Brand v/s Model")
plt.savefig('model.png',bbox_inches='tight')
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()
x=data.groupby('Brand')['Model'].count().index
y=data.groupby('Brand')['Model'].count().values
ax.bar(x,y,color=[color_generate() for i in range(len(x))])
plt.xticks(rotation=85)
plt.show()
```

```python
d={}
for i,j in data.groupby('Brand')['Model'].value_counts().index:
    if i not in d:
        l=[]
        l.append(j)
        d.update({i:l})
    else:
        d[i].append(j)
    
    
```
